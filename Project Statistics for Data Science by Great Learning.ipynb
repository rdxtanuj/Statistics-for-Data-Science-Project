```python
#Import pandas
import pandas as pd
```

```python
#Load data from a CSV file
data=pd.read_csv("Laliga2.csv")
```

# Explore the Data

```python
#Display the first few rows of the DataFrame
data.head(10)
```

```python
#Get basic statistics of the numeric columns
data.describe(include="all")
```

```python
#Get the shape of the DataFrame
data.shape
```

```python
#Check for missing values
data.isnull().sum()
```

# Task 1: Replace dashes with 0

```python
data.replace('-', 0, inplace=True)
```

# Task 2: Print teams that started playing between 1930-1980

```python
#Step 1: Filter the DataFrame based on debut years
teams_between_1930_1980 = data[(data['Debut'] >= '1930') & (data['Debut'] <= '1980')]
```

```python
#Step 2: Print the results
print("Teams that started playing between 1930-1980:")
print(teams_between_1930_1980['Team'])
```

# Task 3: Print top 5 teams in terms of points

```python
#Step 1: Convert the 'Points' column to numeric
data['Points'] = pd.to_numeric(data['Points'], errors='coerce')
```

```python
#Step 2: Sort the DataFrame by 'Points' in descending order
top_5_teams = data.sort_values(by='Points', ascending=False).head(5)
```

```python
#Step 3: Print the results
print("\nTop 5 teams in terms of points:")
print(top_5_teams[['Team', 'Points']])
```

# Task 4: Goal_diff_count function and finding max and min goal difference teams

```python
#Step 1: Define the Goal_diff_count function
def Goal_diff_count(row):
    return int(row['GoalsFor']) - int(row['GoalsAgainst'])
```

```python
#Step 2: Apply the Goal_diff_count function to create a new column 'GoalDiff'
data['GoalDiff'] = data.apply(Goal_diff_count, axis=1)
```

```python
#Step 3: Find the team with the maximum goal difference
max_goal_diff_team = data.loc[data['GoalDiff'].idxmax()]
```

```python
#Step 4: Find the team with the minimum goal difference
min_goal_diff_team = data.loc[data['GoalDiff'].idxmin()]
```

```python
#Step 5: Print the results for maximum goal difference
print("\nTeam with maximum goal difference:")
print(max_goal_diff_team[['Team', 'GoalDiff']])
```

```python
#Step 6: Print the results for minimum goal difference
print("\nTeam with minimum goal difference:")
print(min_goal_diff_team[['Team', 'GoalDiff']])
```

# Task 5: Create a new column "Winning Percent" and print top 5 teams with the highest winning percentage

```python
#Step 1: Calculate Winning Percentage
data['Winning Percent'] = (pd.to_numeric(data['GamesWon'], errors='coerce') / pd.to_numeric(data['GamesPlayed'], errors='coerce')) * 100
```

```python
#Step 2: Fill NaN values with 0
data['Winning Percent'].fillna(0, inplace=True)
```

```python
#Step 3: Sort the DataFrame by 'Winning Percent' in descending order
top_5_winning_percent_teams = data.sort_values(by='Winning Percent', ascending=False).head(5)
```

```python
#Step 4: Print the results
print("\nTop 5 teams with the highest winning percentage:")
print(top_5_winning_percent_teams[['Team', 'Winning Percent']])
```

# Task 6: Group teams based on their "Best position" and print the sum of their points

```python
#Step 1: Group by 'BestPosition' and sum the 'Points'
grouped_data = data.groupby('BestPosition')['Points'].sum().reset_index().head(10)
```

```python
#Step 2: Print the results
print("\nSum of points for each Best Position:")
print(grouped_data)
```
